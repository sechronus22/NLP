- Extract all NOUN tokens from this story. You will
have to read the file in python first to collect all the text and then extract NOUNs in a python list
- Extract all numbers (NUM POS type) in a python list
- Print a count of all POS tags in this story

```python
cd /content/drive/MyDrive/Study/NLP/codebasics/7. POS
```

```python
with open('news_story.txt') as f:
  text = f.read()

print(text)
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp(text)

print([token.pos_ for token in doc[:10]])
```

```python
noun_list = []
number_list = []

for token in doc:
  if token.pos_ == 'NOUN':
    noun_list.append(token.text)
  elif token.pos_ == 'NUM':
    number_list.append(token.text)

print(noun_list[:10])
print(number_list[:10])
```

```python
# my solution
from collections import Counter

tag_list = [token.tag_ for token in doc]
tag_count = Counter(tag_list)
for k,v in tag_count.items():
  print(f'{k} : {v}')
```

```python
count = doc.count_by(spacy.attrs.POS)
for k,v in count.items():
    print(doc.vocab[k].text, "|",v)
```

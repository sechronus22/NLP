```python
import gensim.downloader as api

wv = api.load("word2vec-google-news-300")
```

```python
wv.similarity(w1="profit", w2='well')
```

```python
wv.similarity(w1="profit", w2='gain')
```

```python
wv.most_similar("good")
```

Note that similiarity in word embedding means 2 words usually appear in similar context.

```python
wv.most_similar("dog")
```

### Word combination

```python
wv.most_similar(positive=['king','woman'], negative=['man'])
```

```python
wv.most_similar(positive=['postman','delivery'], negative=['human'])
```

```python
wv.most_similar(positive=['france', 'berlin'], negative=['paris'], topn=5)
```

```python
wv.doesnt_match(['dog','cat','lion','house'])
```

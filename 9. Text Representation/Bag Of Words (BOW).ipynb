## Bag Of Words (BOW)
- use cases : use BOW to create count vectorizer and use them to train spam classification model

```python
import pandas as pd
import numpy as np
```

```python
cd /content/drive/MyDrive/Study/NLP/codebasics/9. Text Representation
```

```python
df = pd.read_csv("spam.csv")
df.head()
```

```python
df['Category'].value_counts()
```

```python
df['spam'] = df.Category.apply(lambda x : 1 if x == 'spam' else 0)
df.head()
```

## Train test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df['Message'], df['spam'],test_size=0.2, random_state=40)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

## Create bag of words representation using CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()

X_train_cv = cv.fit_transform(X_train.values)
X_train_cv
```

```python
X_train_np = X_train_cv.toarray()
X_train_np[0]
```

```python
cv.get_feature_names_out()[6977]
```

```python
dir(cv)
```

```python
cv.vocabulary_ # show vocab and its index
```

```python
# Explore how sparse X_train is
np.where(X_train_np[0] != 0)
```

```python
X_train.iloc[0]
```

## Train the naive bayes model

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train_cv, y_train)
```

```python
X_test_cv = cv.transform(X_test)
```

## Evaluate performance

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test_cv)
print(classification_report(y_test, y_pred))
```

```python
emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]

emails_count = cv.transform(emails)
model.predict(emails_count)
```

## Train the model using sklearn pipeline and reduce number of lines of code

```python
from sklearn.pipeline import Pipeline

clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)

print(classification_report(y_test, y_pred))
```
